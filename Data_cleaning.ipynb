# Data cleaning

```python
import pandas as pd

# suppress copy warning
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# import data
raw_df = pd.read_csv("yelp_academic_dataset_business.csv")
raw_df.head()
```

```python
# many instances have missing value in "neighborhood"
len(raw_df[raw_df.neighborhood.notnull()])
```

```python
# remove neighborhood attribute since it has too many missing values 
raw_df.drop(raw_df.columns[[0]], axis=1, inplace=True)
raw_df.head()
```

```python
len(raw_df)
```

```python
# remove instances with missing values in attributes
raw_df = raw_df.dropna()
len(raw_df)
```

```python
# extract attributes with aggregated values
subset_raw_df = raw_df[['hours','attributes','categories']]
subset_raw_df.head()
```

```python
# clean hours
subset_raw_df['hours'] = subset_raw_df['hours'].map(lambda x:x.strip('[').strip(']').strip("u'").replace("', u'",','))
subset_raw_df['hours'].head()
```

```python
# function to split hours
def splitHours(x):
    d = {"Monday":'N', "Tuesday":'N', "Wednesday":'N', "Thursday":'N', "Friday":'N', "Saturday":'N', "Sunday":'N'}
    hourList = x.split(',')
    for hour in hourList:
        components = hour.split(' ')
        for day in d.keys():
            if day == components[0]:
                d[day] = components[1]
                break
    return d            
```

```python
# split hours
hours_dict_df = subset_raw_df['hours'].apply(lambda x: splitHours(x))
hours_dict_df.head()
```

```python
# convert it into a feature matrix hours_df
hours_df = hours_dict_df.apply(pd.Series)
hours_df.head()
```

```python
# merge raw_df and hours_df
merge_df = pd.merge(raw_df, hours_df, right_index=True, left_index=True)
merge_df.head()
```

```python
# remove non-character symbols in categories
subset_raw_df['categories'] = subset_raw_df['categories'].map(lambda x:x.strip('[').strip("']").strip("\"]").replace("', u'",',').replace("\", u'",',').replace("', u\"",',').replace("\", u\"",',').strip("u'").strip("u\""))
subset_raw_df['categories'].head()
```

```python
# function to identify restaurant from all categories
def isRestaurant(x):
    d = "N"
    categories = x.split(',')
    for category in categories:
        if category == "Restaurants":
            d = "Y"
            break     
    return d  
```

```python
# merge merge_df with a new attribute "isRestaurant"
merge_df['categories'] = subset_raw_df['categories']
merge_df['isRestaurant'] = merge_df['categories'].apply(lambda x: isRestaurant(x)) 
merge_df.head()
```

```python
# remove instances not belonging to restaurant category
restaurant_df = merge_df[merge_df.isRestaurant == 'Y']
restaurant_df.reset_index(drop=True,inplace=True)
restaurant_df.head()
```

```python
# function to split categories
def splitCategories(x):
    d = {}
    categories = x.split(',')
    for category in categories:
        d[category] = "Y"
    return d  
```

```python
# split categories
categories_dict_df = restaurant_df['categories'].apply(lambda x: splitCategories(x)) 
categories_dict_df.head()
```

```python
# convert it into a feature matrix categories_df
categories_df = pd.DataFrame.from_records(categories_dict_df.tolist())
categories_df.fillna("N", inplace=True)
categories_df.head()
```

```python
# merge the dataframes
restaurant_df = pd.merge(restaurant_df, categories_df, right_index=True, left_index=True)
restaurant_df.head()
```

```python
# remove non-character symbols in categories
restaurant_df['attributes'] = restaurant_df['attributes'].map(lambda x:x.strip('[').strip(']').strip("u'").replace("}\", u'",',').replace(": {",' ').replace("', u'",',').replace("', u\"",',').replace("'","").strip("}\"").strip("u\""))
restaurant_df['attributes'].head()
```

```python
# function to split attributes
def splitAttributes(x):
    d = {}
    attributeList = x.split(',')
    for attribute in attributeList:
        components = attribute.split(':')
        d[components[0].strip()] = components[1].strip()
        
    return d  
```

```python
# split attributes
attributes_dict_df = restaurant_df['attributes'].apply(lambda x: splitAttributes(x)) 
attributes_dict_df.head()
```

```python
# transform into a feature matrix attributes_df
attributes_df = pd.DataFrame.from_records(attributes_dict_df.tolist())
attributes_df.fillna("Missing", inplace=True)
attributes_df.head()
```

```python
# merge dataframes
restaurant_df = pd.merge(restaurant_df, attributes_df, right_index=True, left_index=True)
restaurant_df.head()
```

```python
# remove the original "hours", "categories", and "attributes" attributes
restaurant_df.drop(raw_df.columns[[1,4,5]], axis=1, inplace=True)
restaurant_df.head()
```

```python
# remove all non-character symbols in data
import re
pattern = re.compile('[^a-zA-Z]')
restaurant_df.rename(columns=lambda x: pattern.sub('', x), inplace=True)
restaurant_df.head()
```

```python
# remove duplicated columns if any
restaurant_no_duplicate_df = restaurant_df.loc[:,~restaurant_df.columns.duplicated()]
```

```python
# replace commas in data with _, otherwise it may cause problems when exporting into a csv file
restaurant_no_duplicate_df = restaurant_no_duplicate_df.apply(lambda x: x.replace(',','_'))
restaurant_no_duplicate_df.head()
```

```python
# export data into csv
restaurant_no_duplicate_df.to_csv('restaurants.csv', sep=',', encoding='utf-8', index=False)
```

```python
# summary of the review count attribute
restaurant_no_duplicate_df['reviewcount'].describe()
```

```python
# remove outliers for review count
r_n_df = restaurant_no_duplicate_df[((restaurant_no_duplicate_df.reviewcount-restaurant_no_duplicate_df.reviewcount.mean()).abs()/restaurant_no_duplicate_df.reviewcount.std()) < 3]
```

```python
# summary of review count after removing outliers
r_n_df['reviewcount'].describe()
```

```python
# normalize review count
r_n_df['reviewcount'] = r_n_df['reviewcount'].map(lambda x: (x-3)/(550-3))
```

```python
# summary of review count after normalization
r_n_df['reviewcount'].describe()
```

```python
# summary of star count
r_n_df['stars'].describe()
```

```python
# normalize star count
r_n_df['stars'] = r_n_df['stars'].map(lambda x: (x-1)/(5-1))
r_n_df['stars'].describe()
```

```python
# create a new dataframe "popularity_df" by multiplying normalized star count and normalized review count
popularity_df = r_n_df['reviewcount'].multiply(r_n_df['stars'], axis="index") 
popularity_df.describe()
```

```python
# normalize popularity
popularity_df = popularity_df.map(lambda x: (x-0)/(0.877514-0))
popularity_df.describe()
```

```python
# create a new attribute "popularity" and assign data to it
popularity_df = popularity_df.to_frame("Popularity")
r_n_df = pd.merge(r_n_df, popularity_df, right_index=True, left_index=True)
r_n_df.head()
```

```python
# export the normalized data into a csv file
r_n_df.to_csv('restaurants_norm.csv', sep=',', encoding='utf-8', index=False)
```
